```python
import numpy as np
from scipy.optimize import newton

# Definir la función f(x) = x^(1/3)
def f(x):
    return np.cbrt(x)  # raíz cúbica

# Derivada: f'(x) = (1/3) * x^(-2/3)
def df(x):
    return (1/3) * x**(-2/3) if x != 0 else np.inf

```

### Método de Newton (con derivada), x0 = 1

```python
print("Método de Newton (x0 = 1):")
try:
    raiz_newton = newton(f, x0=1.0, fprime=df, maxiter=10, tol=1e-8, full_output=True, disp=True)
    print(f"  Raíz ≈ {raiz_newton[0]:.10f}")
except Exception as e:
    print(f"  Error en método de Newton: {e}")
```

### Método de la secante (sin derivada), p0 = 5, p1 = 0.5

```python
print("\nMétodo de la Secante (p0 = 5, p1 = 0.5):")
try:
    raiz_secante = newton(f, x0=5.0, x1=0.5, maxiter=10, tol=1e-8, full_output=True, disp=True)
    print(f"  Raíz ≈ {raiz_secante[0]:.10f}")
except Exception as e:
    print(f"  Error en método de la secante: {e}")
```
