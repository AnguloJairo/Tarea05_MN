```python
import numpy as np
from scipy.optimize import newton

def f(x):
    return -x**3 - np.cos(x)

def df(x):
    return -3*x**2 + np.sin(x)

```

### Método de Newton-Raphson 

```python

print("Método de Newton :")
p0 = -1.0

# Primera iteración
p1 = p0 - f(p0)/df(p0)
print(f"p1 = {p1:.6f}")

# Segunda iteración
p2 = p1 - f(p1)/df(p1)
print(f"p2 = {p2:.6f}")
```

### Método de la Secante 

```python
print("\nMétodo de la Secante:")
p0_sec = -1.0
p1_sec = -1.1

# Primera iteración
p2_sec = p1_sec - f(p1_sec)*(p1_sec - p0_sec)/(f(p1_sec) - f(p0_sec))
print(f"Primera iteración: p2 = {p2_sec:.6f}")

# Segunda iteración
p3_sec = p2_sec - f(p2_sec)*(p2_sec - p1_sec)/(f(p2_sec) - f(p1_sec))
print(f"Segunda iteración: p3 = {p3_sec:.6f}")
```

### Verificación con p0 = 0

```python

print("\nPrueba con p0 = 0:")
try:
    # Newton no funciona en x=0 porque f'(0)=0
    df_at_0 = df(0)
    print(f"f'(0) = {df_at_0:.1f} → División por cero en Newton")
except Exception as e:
    print(f"Error: {e}")

# Para la secante con p0=0, necesitamos p1≠0
print("\nSecante con p0=0 y p1=0.1:")
p0_test = 0.0
p1_test = 0.1
p2_test = p1_test - f(p1_test)*(p1_test - p0_test)/(f(p1_test) - f(p0_test))
print(f"Primera iteración: {p2_test:.6f}")
```
