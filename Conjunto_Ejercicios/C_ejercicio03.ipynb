```python
import numpy as np
from scipy.optimize import newton

# Tolerancia solicitada
tol = 1e-5

```

### a) f(x) = 3x - e^x

```python
def fa(x):
    return 3*x - np.exp(x)

def dfa(x):
    return 3 - np.exp(x)

# Newton
ra_newton = newton(fa, x0=1.5, fprime=dfa, tol=tol)

# Secante 
ra_secante = newton(fa, x0=1.0, x1=2.0, tol=tol)

print("a) f(x) = 3x - e^x")
print(f"  Método de Newton : x ≈ {ra_newton:.6f}")
print(f"  Método de Secante: x ≈ {ra_secante:.6f}")
```

### b) f(x) = 2x + 3cos(x) - e^x

```python
def fb(x):
    return 2*x + 3*np.cos(x) - np.exp(x)

def dfb(x):
    return 2 - 3*np.sin(x) - np.exp(x)

# Newton
rb_newton = newton(fb, x0=1.5, fprime=dfb, tol=tol)

# Secante
rb_secante = newton(fb, x0=1.0, x1=2.0, tol=tol)

print("\nb) f(x) = 2x + 3cos(x) - e^x")
print(f"  Método de Newton : x ≈ {rb_newton:.6f}")
print(f"  Método de Secante: x ≈ {rb_secante:.6f}")
```
