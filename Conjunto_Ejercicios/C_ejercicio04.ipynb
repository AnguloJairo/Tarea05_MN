```python
import numpy as np
from scipy.optimize import newton

# Definir la función y su derivada
def f(x):
    return 230*x**4 + 18*x**3 + 9*x**2 - 221*x - 9

def df(x):
    return 920*x**3 + 54*x**2 + 18*x - 221

# Tolerancia deseada
tol = 1e-6

# --- Raíz en [-1, 0] ---
print("Raíz en [-1, 0]")

# a) Secante (usa extremos)
r1_secante = newton(f, x0=-1, x1=0, tol=tol)
print(f"  a) Secante: x ≈ {r1_secante:.8f}")

# b) Newton (usa punto medio)
r1_newton = newton(f, x0=-0.5, fprime=df, tol=tol)
print(f"  b) Newton : x ≈ {r1_newton:.8f}")

# --- Raíz en [0, 1] ---
print("\nRaíz en [0, 1]")

# a) Secante
r2_secante = newton(f, x0=0, x1=1, tol=tol)
print(f"  a) Secante: x ≈ {r2_secante:.8f}")

# b) Newton
r2_newton = newton(f, x0=0.5, fprime=df, tol=tol)
print(f"  b) Newton : x ≈ {r2_newton:.8f}")
```
