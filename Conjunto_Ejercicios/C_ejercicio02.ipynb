```python
import numpy as np
from scipy.optimize import newton
```

### a. x^3 - 2x^2 - 5 = 0 en [1, 4]

```python
def fa(x):
    return x**3 - 2*x**2 - 5
def dfa(x):
    return 3*x**2 - 4*x

ra_a = newton(fa, x0=1.0, fprime=dfa, tol=1e-4)
print(f"a) Raíz ≈ {ra_a:.6f}")
```

### b. x^3 + 3x^2 - 1 = 0 en [-3, -2]


```python
def fb(x):
    return x**3 + 3*x**2 - 1
def dfb(x):
    return 3*x**2 + 6*x

rb_b = newton(fb, x0=-3.0, fprime=dfb, tol=1e-4)
print(f"b) Raíz ≈ {rb_b:.6f}")
```

### c. x - cos(x) = 0 en [0, π/2]


```python
def fc(x):
    return x - np.cos(x)
def dfc(x):
    return 1 + np.sin(x)

rc_c = newton(fc, x0=0.5, fprime=dfc, tol=1e-4)
print(f"c) Raíz ≈ {rc_c:.6f}")
```

### d. x - 0.8 - 0.2*sin(x) = 0 en [0, π/2]

```python

def fd(x):
    return x - 0.8 - 0.2 * np.sin(x)
def dfd(x):
    return 1 - 0.2 * np.cos(x)

rd_d = newton(fd, x0=0.5, fprime=dfd, tol=1e-4)
print(f"d) Raíz ≈ {rd_d:.6f}")
```
